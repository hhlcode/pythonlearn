```python
import imageio
import matplotlib.pyplot as plt
from PIL import Image

from wordcloud import WordCloud, STOPWORDS
```

```python
s= '枸杞 山药 山楂 菊花 生姜 茯苓 葛根 大枣 甘草 黄精 金银花 蜂蜜 莲子 枸杞子 百合 蒲公英 马齿苋 红枣 黄芪 紫苏 沙棘 决明子 桑叶 当归 陈皮 三七 茶叶 花椒 鱼腥草 荞麦 苦瓜 阿胶 肉桂 桔梗 大蒜 绿豆 薄荷 芡实 芡实 丁香 酸枣仁 木瓜 西洋参 薏苡仁  黑芝麻 桑椹 杏仁 藿香' 
```

```python
mask = imageio.imread('picture.png')
# 绘制词云
my_cloud = WordCloud(
    mask = mask, 
    background_color='white',  # 设置背景颜色  默认是black
    font_path='/home/hhl/anaconda3/lib/python3.8/site-packages/wordcloud/SimHei.ttf',   # 设置字体  显示中文
    max_font_size=112,        # 设置字体最大值
    min_font_size=12,         # 设置字体最小值
    random_state=88           # 设置随机生成状态，即多少种配色方案
).generate(s)
# 绘制词云
plt.figure(figsize=(8, 5), dpi=200)
# 显示生成的词云图片
plt.imshow(my_cloud, interpolation='bilinear')
# 显示设置词云图中无坐标轴
plt.axis('off')
plt.savefig('sample.png', dpi=200)
plt.show()
```
