```python
import pandas as pd
import jieba
import jieba.analyse
df = pd.read_csv('药食同源.csv',encoding ='utf-8')
df
```

```python
abstract = df.摘要
abstract
```

```python
"""
#Python 消除换行符一定.replace('\n','').replace('\r','') 
将词表转化为一个list
"""
with open ('中药大词典合集.txt','r',encoding = 'utf-8')as f:
    vocabL = f.readlines() #读取的结果是list 
    vocabSet = []
    for i in vocabL:#删除换行符
        i = i.replace('\n','').replace('\r','').replace('t','')
        vocabSet.append(i)
    f.close()
#     print(vocabSet)
```

```python
jieba.load_userdict("中药大词典合集.txt")#引进自定义的词表
c = []
for ab in abstract:
    l = [] #把每个文章的用到的药材提取出来放在列表中
#     contents = jieba.analyse.extract_tags(str(ab), topK=40, withWeight=True, allowPOS=('n','nr','ns'))#中文提取分词,目前在这不可行
#     for content in contents:
#         print(content[0])
    contents = list(jieba.cut(str(ab),cut_all=False, HMM=True))#精确模式  隐马尔科夫模型
#         print(contents)
    for content in contents:
#             print(content)
        if content  in vocabSet and content !='药食同源'and content not in l: #判断是否在词表里面，提取中药材的关键词
            l.append(content)
    if len(l)==0:
        c.append('999')
    else: 
        s = str(l).replace("'",'').replace('[','').replace(']','')#将其转化为一个一维的列表目的就是为转化为DataFrame类型
        c.append(s)
    l = []
```

```python
df['引进词表的jieba提取的中药材']= pd.DataFrame(c)
```

```python
df.to_csv('药食同源.csv')
```
