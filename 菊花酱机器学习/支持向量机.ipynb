```python
import numpy  as np
import pandas as pd
"""
 函数功能： 创建特征向量和标签向量
 参数说明： 
 file :原始文件路径
 返回: 
 xMat: 特征向量
 yMat: 标签向量
"""
def loadDataSet(file):
    dataSet = pd.read_table(file,header = None)
    xMat = np.mat(dataSet.iloc[:,:-1].values)
    yMat = np.mat(dataSet.iloc[:,:-1].values).T#T的作用是干嘛
    return  xMat,yMat
```
